```python
import pandas as pd
import re
from tagging import tagging
from collections import defaultdict
```

```python
shop2info = []
df = pd.read_csv('./genderdate.csv')
```

```python
df
```

```python
tup2count = defaultdict(list)
```

```python
tup2money = defaultdict(int)
```

```python
for idx, row in enumerate(df.values):
    info = []
    #info.append(df['id'][idx])
    if df['birthday'][idx] == df['birthday'][idx]:
        y, _, _ = df['birthday'][idx].split('-')
        age = 2020 - int(y)
        if age <= 25:
            age_str = '年輕'
        elif age > 25 and age <= 40:
            age_str = '中年'
        elif age > 40 and age < 65:
            age_str = '壯年'
        else:
            age_str = '老年'
        info.append(df['gender'][idx])
        info.append(age_str)
    else:
        info.append(df['gender'][idx])
        info.append('No age')
    
#     _, m, _ = df['created_date'][idx].split('-')
#     if int(m) > 1 and int(m) <= 4:
#         season_str = 'spring'
#     elif int(m) > 4 and int(m) <= 7:
#         season_str = 'summer'
#     elif int(m) > 7 and int(m) <= 10:
#         season_str = 'autumn'
#     else:
#         season_str = 'winter'
#     info.append(season_str)
    h, _, _ = df['created_time'][idx].split(':')
    if int(h) <= 4:
        time_str = 'midnight'
    elif int(h) > 4 and int(h) <= 10:
        time_str = 'morning'
    elif int(h) > 10 and int(h) <= 13:
        time_str = 'noon'
    elif int(h) > 13 and int(h) <= 17:
        time_str = 'afternoon'
    elif int(h) > 17 and int(h) <= 23:
        time_str = 'evening'
    info.append(time_str)
    tup2count[tuple(info)].append(df['id'][idx])
    tup2money[tuple(info)] += df['total_price'][idx]
    shop2info.append(info)
```

```python
df['info'] = shop2info
```

```python
analysis = defaultdict(list)
for idx, row in enumerate(df.values):
    analysis[df['shop_id'][idx]].append(tuple(df['info'][idx]))
```

```python
len(analysis)
```

```python
new_df = defaultdict(list)
for key in analysis.keys():
    for tup in set(analysis[key]):
        #order_count = analysis[key].count(tup)
        new_tup = (*tup, len(set(tup2count[tup])), len(tup2count[tup]), tup2money[tup])
        new_df[key].append(new_tup)
        del new_tup
```

```python
new_df
```

```python
table = []
for key in new_df.keys():
    for tup in new_df[key]:
        row = []
        row.append(key)
        for ele in tup:
            row.append(ele)
        table.append(row)
```

```python
table = pd.DataFrame(table, columns = ['shop_id', 'gender', 'age', 'time', 'purchased_customer_count', 'total_orders', 'total_price'])
```
